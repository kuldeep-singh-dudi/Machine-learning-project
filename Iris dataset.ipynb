#### importing all the required libraries

#### Importing the dataset from the scikit learn library

```python
from sklearn import datasets
```

```python
data=datasets.load_iris()
```

```python
df=pd.DataFrame(data=np.c_[data['data'],data['target']],columns=data['feature_names']+['target'])
```

#### giving valus to variables x and y where x contains the values of rows and features columns and it is 2d array. whereas y is 1D array which contains target values.

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

### importing train tst split library from sklearn to split dataset into train test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

### importing Logistic regression model as in this paper it is given to perform on the logistic regression model

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

### importing metrices for calculating the accuracy score and visualise the confusion matrix and for getting the classificaton report of the model.

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
